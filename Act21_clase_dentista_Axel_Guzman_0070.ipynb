```python
class Dentista:
    # Método constructor para inicializar atributos
    def __init__(self, nombre, especialidad, cedula):
        self.nombre = nombre
        self.especialidad = especialidad
        self.cedula = cedula
        self.consultorio_abierto = False

    # Método para abrir el consultorio
    def abrir_consultorio(self):
        self.consultorio_abierto = True
        print(f"El Dr./Dra. {self.nombre} ha abierto su consultorio de {self.especialidad}.")

    # Método para mostrar información
    def mostrar_info(self):
        return f"Dentista: {self.nombre}, Especialidad: {self.especialidad}, Cédula: {self.cedula}"

# Creación de instancias (objetos)
dentista1 = Dentista("Manuel Nava", "Ortodoncia", "DENT-0023")
dentista2 = Dentista("Elena Flores", "Odontopediatría", "DENT-0055")

# Llamadas a las funciones para cada objeto dentista 1 y dentista 2
print("dentista 1:")
dentista1.abrir_consultorio()
print(dentista1.mostrar_info())

# Uso de métodos
print("\ndentista 2:")
dentista2.abrir_consultorio()
print(dentista2.mostrar_info())

print("\nprograma terminado")
print("creador de software por Axel Guzman NL= 23")
```
